```python
import sys
sys.path.insert(1, '../src')
import app_constants as const

from keras.models import load_model
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
```

```python
PATH_COVID = '/Users/chinmay/Documents/Workspace/dissertation/data/chest_xrays/COVID/COVID-61.png'
PATH_VIRAL = '/Users/chinmay/Documents/Workspace/dissertation/data/chest_xrays/Viral Pneumonia/Viral Pneumonia-33.png'
PATH_NORMAL = '/Users/chinmay/Documents/Workspace/dissertation/data/chest_xrays/Normal/Normal-52.png'
```

```python
def load_image(img_path):
    print(img_path)    
    img = image.load_img(img_path, color_mode = "rgb", target_size=(256, 256))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    
    plt.imshow(img)                           
    plt.axis('off')
    plt.show()
    
    print(x.shape)

    return x
```

# =============================================================================
# Model 1
# =============================================================================

```python
model_1 = load_model(const.get_model_path("model1"))
model_1.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
```

```python
# load a single image
img = load_image(const.get_random_pneumonia_xray())
prediction_1 = model_1.predict(img)
y_classes_1 = prediction_1.argmax(axis=-1)
print("Class: {},  Label: {}".format(y_classes_1, const.LABELS[np.argmax(prediction_1)]))
```
